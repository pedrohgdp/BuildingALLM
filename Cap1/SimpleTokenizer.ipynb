# Tokenizer

A tokenizer is a class where we take the original text and split it into tokens.

For example, if we have the text:

```text
"Hello, Pedro, how are you?"
```

After passing it to the tokenizer we will get an array like:

```text
["hello", ",", "pedro", ",", "how", "are", "you", "?"]
```

After getting this token array, we can create the token id tensors and the embeddings, which we will see later.

But before that, we also need to create a vocabulary first.

# Vocabulary

Given a token array like the one above, we first sort it in alphabetical order.

Then we create a vocabulary where each token in the array receives an id.

So in this case the vocabulary would look something like:

```python
{
    ",": 0,
    "?": 1,
    "are": 2,
    "hello": 3,
    "how": 4,
    "pedro": 5,
    "you": 6
}
```

With this vocabulary we can create two types of maps.

The text-to-id map and the id-to-text map.

The text-to-id map is the vocabulary itself.

The id-to-text map is created by iterating through the vocabulary and saving the id as the key and the token as the value.

Encoding is the process of passing a text and receiving its token ids.

Decoding is the process of passing token ids and receiving the original text back.

That is what the code below does.

# Handling unknown texts

When we have unknown words, or when we want to represent the end of one document and the start of another, we can use special tokens like:

```text
"<|endoftext|>"
```

for the end of a document, and

```text
"<|unk|>"
```

for words that are not inside the vocabulary.

So we add these special tokens to the vocabulary after tokenizing the text.

# Summary

In the code below, we use the tokenize function both for tokenizing a text and creating the vocabulary, and later for tokenizing new texts before converting them into ids using the vocabulary we created before.

````python
import re
from google.colab import files

uploaded = files.upload()
filename = list(uploaded.keys())[0]

with open(filename, "r", encoding="utf-8") as f:
    text = f.read()


def tokenize(text):
    text = text.lower()

    text = re.sub(r'[^a-z0-9\s,.?!\'"-]', ' ', text)

    text = re.sub(r'\s+', ' ', text)

    tokens = re.split(r'([,.?!\'"-]|\s)', text)

    return [t.strip() for t in tokens if t.strip()]

#Vocabulary creation
tokens = tokenize(text)

unique_tokens = sorted(set(tokens))
unique_tokens.extend(["<|endoftext|>", "<|unk|>"])

vocab = {token: i for i, token in enumerate(unique_tokens)}


print("Texto exemplo:\n", text[:200])
print("\nVocab size:", len(vocab))

class SimpleTokenizer:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_string = {i: s for s, i in vocab.items()}

    def encoded(self, text):
        tokens = tokenize(text)
        return [ self.str_to_int.get(item, self.str_to_int["<|unk|>"]) for item in tokens]

    def decoded(self, ids):
        text = " ".join(self.int_to_string[i] for i in ids)
        text = re.sub(r'\s+([,.?!\'"])', r'\1', text)
        return text


tokenizer = SimpleTokenizer(vocab)

sample = "hello world! this is a test."
encoded = tokenizer.encoded(sample)
decoded = tokenizer.decoded(encoded)

print("\nSample:", sample)
print("Encoded:", encoded)
print("Decoded:", decoded)

print("\nPrimeiros tokens do vocab:")
for token, idx in list(vocab.items())[:50]:
    print(token, idx)
````
